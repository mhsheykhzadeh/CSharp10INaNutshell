<p>وقتی یک برنامه را از یک IDE مثل Visual Studio در حالت دیباگ اجرا می‌کنید، دیباگر به صورت خودکار به برنامه متصل می‌شود. اما در مواردی، شروع برنامه در حالت دیباگ از داخل IDE امکان‌پذیر یا راحت نیست:</p>

<ul><li><p><strong>برنامه‌های خدماتی ویندوز (Windows Service)</strong>: این برنامه‌ها معمولاً توسط سیستم‌عامل اجرا می‌شوند و به صورت دستی از IDE شروع نمی‌شوند.</p></li><li><p><strong>طراح ویژوال استودیو (Visual Studio Designer)</strong>: گاهی لازم است کدی که در محیط طراح ویژوال استودیو اجرا می‌شود را دیباگ کنید.</p></li></ul>

<p>در این موارد، می‌توانید از متد <code>Debugger.Break</code> در کد خود استفاده کنید. وقتی این متد اجرا می‌شود:</p>
<ol><li>دیباگر را راه‌اندازی می‌کند.</li><li>دیباگر را به برنامه در حال اجرا متصل می‌کند.</li><li>اجرای برنامه را در نقطه‌ای که <code>Debugger.Break</code> فراخوانی شده است متوقف می‌کند.</li></ol>

<p>فرض کنید یک Windows Service در حال توسعه دارید و نیاز دارید متد <code>OnStart</code> آن را دیباگ کنید. از آنجایی که نمی‌توانید سرویس را مستقیماً از ویژوال استودیو در حالت دیباگ شروع کنید، می‌توانید <code>Debugger.Break</code> را در ابتدای متد <code>OnStart</code> قرار دهید:</p>
</div>

```python
protected override void OnStart(string[] args)
{
    Debugger.Break();
    // کد راه‌اندازی سرویس شما در اینجا
}
```

<div dir="rtl" style="width:90%; margin:auto">
<p><p><strong>دیباگ در محیط تولید (Production)</strong>: استفاده از <code>Debugger.Break</code> یا <code>Debugger.Launch</code> در محیط تولید باید با احتیاط انجام شود. این ابزارها بیشتر برای محیط توسعه و تست مناسب هستند.</p></p>

<p>
دیباگ شرطی: می‌توانید از عبارات شرطی استفاده کنید تا کد دیباگ تنها در محیط توسعه اجرا شود:</p>
</div>

```python
if (Debugger.IsAttached)
{
    Debugger.Break();
}
```

### Debugger Attributes

<div dir="rtl" style="width:90%; margin:auto">
<p>ویژگی‌های <code>DebuggerStepThrough</code> و <code>DebuggerHidden</code> ابزارهایی هستند که به توسعه‌دهندگان اجازه می‌دهند رفتار دیباگر را کنترل کنند و تجربه دیباگ کردن را بهبود بخشند.</p>

<ul><li><p><strong>DebuggerStepThrough</strong>: وقتی این ویژگی به یک متد، سازنده، یا کلاس اعمال می‌شود، دیباگر از روی کد عبور می‌کند بدون اینکه کاربر مجبور باشد به صورت دستی هر مرحله را تایید کند. این ویژگی در موارد زیر مفید است:</p><ul><li><strong>متدهای تولید شده به صورت خودکار</strong>: مانند کدهایی که توسط ابزارها یا کتابخانه‌ها تولید می‌شوند و شما معمولاً نمی‌خواهید داخل این کدها را دیباگ کنید.</li><li><strong>متدهای پراکسی</strong>: متدهایی که به جای اجرای منطق خود، کار را به متدهای دیگری ارجاع می‌دهند. در این حالت، عبور از روی متد پراکسی و تمرکز بر روی متد واقعی مفید است.</li></ul></li><li><p><strong>DebuggerHidden</strong>: این ویژگی به دیباگر دستور می‌دهد که کد مشخص شده را کاملاً پنهان کند. وقتی این ویژگی اعمال می‌شود:</p><ul><li>دیباگر به هیچ وجه به کد وارد نمی‌شود، حتی اگر نقطه شکستی در آن تنظیم شده باشد.</li><li>این ویژگی می‌تواند همراه با <code>DebuggerStepThrough</code> استفاده شود تا کد پراکسی به طور کامل از دید دیباگر مخفی شود و کاربر تنها منطق اصلی برنامه را ببیند.</li></ul></li></ul>

</div>

```python
[DebuggerStepThrough] /* دیباگر از روی آن عبور می‌کند بدون اینکه متوقف شود.*/
public void AutoGeneratedMethod()
{
    // کد تولید شده به صورت خودکار
}

[DebuggerStepThrough, DebuggerHidden] 
/*دیباگر نه تنها از روی آن عبور می‌کند، بلکه به طور کامل آن را نادیده می‌گیرد و تنها بر روی RealMethod تمرکز می‌کند.*/
public void ProxyMethod()
{
    RealMethod();
}

public void RealMethod()
{
    // منطق اصلی برنامه
}
```
