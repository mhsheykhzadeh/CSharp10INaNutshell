<p>در یک برنامه معمولی، علاوه بر کد اجرایی (مثل فایل‌های DLL یا EXE)، ممکن است داده‌های دیگری نیز وجود داشته باشد که بخشی از برنامه هستند. این داده‌ها می‌توانند شامل موارد زیر باشند:</p>
<ul><li><strong>تصاویر (Images)</strong>: مانند فایل‌های JPEG یا PNG.</li><li><strong>متن‌ها (Text)</strong>: که ممکن است نیاز به ترجمه در برنامه‌های چندزبانه داشته باشند.</li><li><strong>فایل‌های XML</strong>: که برای ذخیره‌سازی داده‌های ساخت‌یافته استفاده می‌شوند.</li></ul>

<p><strong>کاربرد منابع</strong>:</p>
<ol><li><strong>ذخیره‌سازی داده‌هایی که نمی‌توانند در کد منبع باشند</strong>: مثل تصاویر یا فایل‌های باینری.</li><li><strong>ذخیره‌سازی داده‌هایی که ممکن است نیاز به ترجمه داشته باشند</strong>: مانند متن‌های مربوط به زبان‌های مختلف در برنامه‌های چندزبانه.</li></ol>

<h5>مفهوم فایل‌های .resources:</h5>
<ul><li><p><strong>فایل‌های .resources</strong>: این فایل‌ها کانتینرهایی هستند که برای نگهداری محتوای قابل محلی‌سازی (مانند رشته‌های متنی، تصاویر، و غیره) طراحی شده‌اند. این محتوا می‌تواند در زبان‌ها یا فرهنگ‌های مختلف باشد و بسته به زبان سیستم‌عامل کاربر در زمان اجرا، استفاده شود.</p></li><li><p><strong>کاربرد</strong>: وقتی شما برنامه‌ای را می‌سازید که نیاز به پشتیبانی از زبان‌های مختلف دارد، فایل‌های .resources به شما کمک می‌کنند تا محتوای مختلف را برای زبان‌های مختلف در فایل‌های جداگانه ذخیره کنید.</p></li></ul>

<h5>اسمبلی‌های ماهواره‌ای (Satellite Assemblies):</h5>
<ul><li><p><strong>تعریف</strong>: اسمبلی‌های ماهواره‌ای نوع خاصی از اسمبلی‌ها هستند که فقط حاوی منابع (resources) محلی‌سازی شده برای یک زبان یا فرهنگ خاص هستند. این اسمبلی‌ها به طور جداگانه از اسمبلی اصلی برنامه تولید می‌شوند.</p></li><li><p><strong>کاربرد</strong>: زمانی که برنامه در محیطی با زبان یا فرهنگ خاص اجرا می‌شود، دات‌نت به طور خودکار اسمبلی ماهواره‌ای مناسب را بر اساس زبان سیستم‌عامل کاربر بارگذاری می‌کند. این فرآیند باعث می‌شود که برنامه به‌صورت خودکار و بدون نیاز به تغییر کد اصلی، محتوای مناسب زبان یا فرهنگ کاربر را نمایش دهد.</p></li></ul>

<h5>چگونگی کارکرد:</h5>
<ol><li><p><strong>ساخت فایل‌های .resources</strong>:</p><ul><li>برای هر زبان یا فرهنگ، یک فایل <code>.resx</code> (فایل XML که منابع را نگه می‌دارد) ایجاد می‌کنید. این فایل‌ها سپس به فایل‌های باینری <code>.resources</code> تبدیل می‌شوند.</li><li>مثال: ممکن است یک فایل <code>Strings.en-US.resx</code> برای زبان انگلیسی و یک فایل <code>Strings.fr-FR.resx</code> برای زبان فرانسوی داشته باشید.</li></ul></li><li><p><strong>ساخت اسمبلی‌های ماهواره‌ای</strong>:</p><ul><li>فایل‌های <code>.resources</code> به اسمبلی‌های ماهواره‌ای تبدیل می‌شوند. هر اسمبلی ماهواره‌ای معمولاً برای یک زبان یا فرهنگ خاص است.</li><li>مثال: یک اسمبلی ماهواره‌ای ممکن است فقط حاوی منابع برای زبان فرانسوی باشد (<code>fr-FR</code>).</li></ul></li><li><p><strong>زمان اجرا (Runtime)</strong>:</p><ul><li>هنگامی که برنامه اجرا می‌شود، دات‌نت به زبان یا فرهنگ سیستم‌عامل کاربر نگاه می‌کند.</li><li>سپس به‌طور خودکار اسمبلی ماهواره‌ای مربوطه را بارگذاری می‌کند تا محتوای مناسب زبان کاربر را نمایش دهد.</li></ul></li></ol>

<p>برای اضافه کردن یک تصویر به پروژه به طوری که به عنوان یک <strong>Resource</strong> در فایل نهایی اسمبلی قرار گیرد، باید مراحل زیر را طی کنید:</p>

<h5>1. <strong>اضافه کردن تصویر به پروژه:</strong></h5>
<ul><li>ابتدا تصویر مورد نظر خود را به پروژه اضافه کنید:<ul><li>در Visual Studio یا محیط توسعه خود، به پروژه‌تان بروید.</li><li>روی پروژه راست کلیک کنید و گزینه‌ی <strong>"Add"</strong> سپس <strong>"Existing Item..."</strong> را انتخاب کنید.</li><li>تصویر مورد نظر (مثلاً <code>image.jpg</code>) را انتخاب کنید و آن را به پروژه اضافه کنید.</li></ul></li></ul>
<h5>2. <strong>تنظیم خصوصیات فایل:</strong></h5>
<ul><li>پس از اضافه کردن تصویر به پروژه، باید تنظیمات آن را تغییر دهید تا به عنوان Resource در فایل اسمبلی قرار گیرد:<ul><li>روی تصویر اضافه شده در Solution Explorer راست کلیک کنید و گزینه‌ی <strong>"Properties"</strong> را انتخاب کنید.</li><li>در پنجره‌ی Properties، خصوصیت <strong>"Build Action"</strong> را به <strong>"Embedded Resource"</strong> تغییر دهید.</li></ul></li></ul>
<h5>3. <strong>کامپایل پروژه:</strong></h5>
<ul><li>حال که تصویر به عنوان یک Embedded Resource تنظیم شده است، پروژه را کامپایل کنید.</li><li>تصویر به صورت بایتی در فایل نهایی اسمبلی (مثلاً <code>YourAssembly.dll</code>) جاسازی خواهد شد.</li></ul>
<h5>4. <strong>دسترسی به تصویر در زمان اجرا:</strong></h5>
<ul><li>برای دسترسی به تصویر جاسازی شده در کد خود، می‌توانید از فضای نام <code>System.Reflection</code> و <code>System.IO</code> برای بارگذاری تصویر استفاده کنید.</li><li>در اینجا یک مثال برای دسترسی به تصویر جاسازی شده آمده است:</li></ul>
</div>

```python
using System;
using System.IO;
using System.Reflection;

public class Program
{
    public static void Main()
    {
        // نام اسمبلی کنونی را دریافت کنید
        Assembly assembly = Assembly.GetExecutingAssembly();

        // نام کامل resource را مشخص کنید (فضای نام.نام فایل)
        string resourceName = "YourNamespace.image.jpg";

        // با استفاده از جریان داده‌ای (Stream) به منبع جاسازی شده دسترسی پیدا کنید
        using (Stream stream = assembly.GetManifestResourceStream(resourceName))
        {
            if (stream != null)
            {
                // داده‌ها را از stream بخوانید (مثلاً تصویر را در حافظه بارگذاری کنید)
                byte[] imageData = new byte[stream.Length];
                stream.Read(imageData, 0, imageData.Length);

                // این داده‌ها را می‌توانید برای بارگذاری تصویر یا استفاده‌های دیگر استفاده کنید
                Console.WriteLine("Image loaded successfully!");
            }
            else
            {
                Console.WriteLine("Resource not found.");
            }
        }
    }
}
```
