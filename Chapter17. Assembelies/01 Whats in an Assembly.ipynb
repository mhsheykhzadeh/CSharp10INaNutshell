<h5>اسمبلی چیست؟</h5>
<ul><li><strong>واحد اصلی استقرار (Deployment)</strong>: اسمبلی‌ها واحدهای اصلی استقرار در .NET هستند. به این معنی که وقتی یک برنامه .NET را منتشر می‌کنید، در قالب اسمبلی‌ها منتشر می‌شود.</li><li><strong>محل نگهداری انواع (Types)</strong>: تمام انواع (کلاس‌ها، اینترفیس‌ها و ...) که در برنامه شما تعریف شده‌اند، در داخل اسمبلی‌ها قرار دارند.</li></ul>

<h5>محتوای یک اسمبلی</h5>
<ul><li><strong>کد زبان واسط (IL)</strong>: اسمبلی‌ها شامل انواع کامپایل شده با کد IL (Intermediate Language) هستند. کد IL یک زبان میانی است که توسط کامپایلر .NET تولید می‌شود و سپس توسط JIT (Just-In-Time) کامپایلر در زمان اجرا به کد ماشین تبدیل می‌شود.</li><li><strong>منابع زمان اجرا (Runtime Resources)</strong>: اسمبلی‌ها ممکن است شامل منابعی مانند تصاویر، فایل‌های متنی و دیگر منابعی باشند که برنامه شما در زمان اجرا به آنها نیاز دارد.</li><li><strong>اطلاعات نسخه‌بندی و ارجاع‌دهی</strong>: اسمبلی‌ها حاوی اطلاعاتی برای کمک به نسخه‌بندی و ارجاع‌دهی به سایر اسمبلی‌ها هستند. این اطلاعات کمک می‌کنند تا سیستم بداند کدام نسخه از اسمبلی‌ها باید استفاده شوند و چگونه اسمبلی‌ها به یکدیگر وابسته‌اند.</li></ul>

<h5>مرز برای حل نوع (Type Resolution)</h5>
<ul><li><strong>مرز برای حل نوع</strong>: اسمبلی‌ها یک مرز تعریف می‌کنند برای اینکه چگونه انواع (Types) شناسایی و بارگذاری شوند. این به این معناست که هر اسمبلی به طور مستقل انواع خود را مدیریت می‌کند و تنها در صورت نیاز به انواع سایر اسمبلی‌ها ارجاع می‌دهد.</li></ul>

### What’s in an Assembly

An assembly contains ***four kinds of things***:  
- An `assembly manifest` : Provides **information** to the `CLR`, such as the **assembly’s name**, **version**, an **other assemblies that it references**
- An `application manifest` : Provides **information** to the `operating system`, such as **how the assembly should be deployed** and whether **administrative elevation is required?**
- `Compiled types` : The **compiled IL code** and **metadata** of the **types defined within the assembly**
- `Resources` : ***Other data embedded*** within the assembly, such as **images** and **localizable text**

### The Assembly Manifest

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">
<h5>اهداف Assembly Manifest</h5>
<ol><li><strong>توصیف اسمبلی برای محیط میزبانی مدیریت شده</strong>: Assembly Manifest اطلاعاتی را فراهم می‌کند که محیط میزبانی مدیریت شده (Managed Hosting Environment) می‌تواند از آن برای شناخت و استفاده از اسمبلی استفاده کند.</li><li><strong>عمل به عنوان دایرکتوری برای ماژول‌ها، انواع، و منابع موجود در اسمبلی</strong>: Assembly Manifest نقش یک دایرکتوری را ایفا می‌کند که شامل اطلاعاتی درباره ماژول‌ها، انواع (Types)، و منابع (Resources) موجود در اسمبلی است.</li></ol>

<h5>خود توصیفی (Self-Describing)</h5>
<ul><li><strong>خود توصیفی بودن اسمبلی‌ها</strong>: اسمبلی‌ها به طور خودکار تمام اطلاعات مورد نیاز را درون خود دارند. این بدان معناست که مصرف‌کنندگان (مانند برنامه‌های دیگر) می‌توانند تمام داده‌ها، انواع، و توابع اسمبلی را بدون نیاز به فایل‌های اضافی کشف کنند.</li></ul>

<h5>داده‌های مهم ذخیره شده در Assembly Manifest</h5>
<ol><li><strong>نام ساده اسمبلی</strong>: نام اصلی اسمبلی.</li><li><strong>شماره نسخه (AssemblyVersion)</strong>: نسخه‌ای که به اسمبلی اختصاص داده شده است.</li><li><strong>کلید عمومی و هش امضای اسمبلی (در صورت قوی‌نام بودن)</strong>: اگر اسمبلی به صورت قوی‌نام (Strongly Named) باشد، شامل کلید عمومی و هش امضای آن است.</li><li><strong>لیست اسمبلی‌های ارجاع شده</strong>: لیستی از اسمبلی‌های دیگری که این اسمبلی به آنها ارجاع داده است، شامل نسخه و کلید عمومی آنها.</li><li><strong>لیست انواع تعریف شده در اسمبلی</strong>: لیستی از انواع (Types) که در این اسمبلی تعریف شده‌اند.</li><li><strong>فرهنگ هدف (اگر اسمبلی ماهواره‌ای باشد) (AssemblyCulture)</strong>: فرهنگی که اسمبلی ماهواره‌ای (Satellite Assembly) به آن اختصاص داده شده است.</li></ol>

<h3>داده‌های اطلاعاتی اضافی در Assembly Manifest</h3>
<ol><li><strong>عنوان و توضیحات کامل (AssemblyTitle و AssemblyDescription)</strong>: عنوان و توضیحات کامل اسمبلی.</li><li><strong>اطلاعات شرکت و حق تکثیر (AssemblyCompany و AssemblyCopyright)</strong>: اطلاعات مربوط به شرکت سازنده و حق تکثیر.</li><li><strong>نسخه نمایش (AssemblyInformationalVersion)</strong>: نسخه‌ای که برای نمایش استفاده می‌شود.</li><li><strong>ویژگی‌های اضافی برای داده‌های سفارشی</strong>: ویژگی‌های اضافی که می‌توانند برای ذخیره داده‌های سفارشی استفاده شوند.</li></ol>

<h5>منبع داده‌های Assembly Manifest</h5>
<ul><li><strong>داده‌های مشتق شده از آرگومان‌های کامپایلر</strong>: برخی از این داده‌ها از آرگومان‌هایی که به کامپایلر داده شده‌اند مشتق می‌شوند، مانند لیست اسمبلی‌های ارجاع شده یا کلید عمومی برای امضای اسمبلی.</li><li><strong>ویژگی‌های اسمبلی</strong>: بقیه داده‌ها از ویژگی‌های اسمبلی (Assembly Attributes) که در کد با پرانتز نشان داده شده‌اند، مشتق می‌شوند.</li></ul>

<p>این کد ویژگی‌های مختلفی را برای اسمبلی تعریف می‌کند که در نهایت در Assembly Manifest ذخیره می‌شوند.</p>
</div>

```python
using System.Reflection;

[assembly: AssemblyTitle("My Assembly")]
[assembly: AssemblyDescription("This is a sample assembly.")]
[assembly: AssemblyCompany("My Company")]
[assembly: AssemblyCopyright("Copyright © My Company 2024")]
[assembly: AssemblyVersion("1.0.0.0")]
[assembly: AssemblyInformationalVersion("1.0.0.0")]
```

### The Application Manifest (Windows)

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">

<ol><li><p><strong>مانیفست برنامه چیست؟</strong></p><ul><li>مانیفست برنامه یک فایل XML است که اطلاعاتی در مورد اسمبلی (مثل نوع و نسخه آن) را به سیستم عامل ویندوز منتقل می‌کند.</li></ul></li><li><p><strong>تعبیه شدن در فایل اجرایی</strong>:</p><ul><li>این مانیفست به عنوان یک منبع Win32 در فایل اجرایی (EXE) برنامه در طی فرایند ساخت برنامه (build process) جاسازی می‌شود. به عبارت دیگر، مانیفست در فایل اجرایی قرار می‌گیرد تا ویندوز بتواند هنگام اجرا به آن دسترسی داشته باشد.</li></ul></li><li><p><strong>نحوه خواندن و پردازش مانیفست</strong>:</p><ul><li>اگر مانیفست موجود باشد، ویندوز قبل از اینکه CLR (Common Language Runtime) اسمبلی را بارگذاری کند، آن را خوانده و پردازش می‌کند. این به این معناست که مانیفست می‌تواند بر نحوه راه‌اندازی و اجرای برنامه توسط ویندوز تأثیر بگذارد.</li></ul></li></ol>
</div>

```python
<?xml version="1.0" encoding="utf-8"?>
<assembly manifestVersion="1.0" xmlns="urn:schemas-microsoft-com:asm.v1">
    <!-- contents of manifest -->
</assembly>
```

The following manifest instructs the OS to ***request administrative elevation***:

<?xml version="1.0" encoding="utf-8"?>
<assembly manifestVersion="1.0" xmlns="urn:schemas-microsoft-com:asm.v1">
    <trustInfo xmlns="urn:schemas-microsoft-com:asm.v2">
        <security>
            <requestedPrivileges>
                <requestedExecutionLevel level="requireAdministrator" />
            </requestedPrivileges>
        </security>
    </trustInfo>
</assembly>

#### Deploying an application manifest

You can **add an application manifest** to a .NET project in Visual Studio by `rightclicking` your **project** in Solution Explorer, selecting Add, then **New Item**, and then choosing `Application Manifest File`. Upon building, the manifest will be embedded into the output assembly.

### The Assembly Class

The `Assembly` class in `System.Reflection` is a gateway to ***accessing assembly metadata*** at `runtime`.

#### How get assembly object?

```python
using System.Reflection;

// simplest way
Assembly a1 = typeof(string).Assembly;

//Returns the assembly of the type that defines the currently executing function
Assembly executingAssembly = Assembly.GetExecutingAssembly();

//Does the same as GetExecutingAssembly but for the function that called the
//currently executing function
Assembly callingAssembly = Assembly.GetCallingAssembly();

//Returns the assembly defining the application’s original entry method
Assembly entryAssembly = Assembly.GetEntryAssembly();
```

After you have an `Assembly object`, you can use its properties and methods to query
the ***assembly’s metadata*** and ***reflect*** upon its types.
