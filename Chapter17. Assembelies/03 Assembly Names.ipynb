<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">

<p> هر اسمبلی در .NET شامل چهار قطعه اطلاعات متادیتا از manifest (مانيفست) خود است که هویت آن را تعیین می‌کند. این چهار جزء عبارتند از:</p>
<ol>
<li><p><strong>نام ساده (Simple Name)</strong>:</p><ul><li>نام ساده از نام فایلی که اسمبلی به آن کامپایل شده است می‌آید (بدون پسوند فایل). به عنوان مثال، نام ساده اسمبلی System.Xml.dll، "System.Xml" است. تغییر نام فایل، نام ساده اسمبلی را تغییر نمی‌دهد.</li></ul></li>
<li><p><strong>نسخه (Version)</strong>:</p><ul><li>شماره نسخه از ویژگی AssemblyVersion به دست می‌آید. این شماره به چهار بخش تقسیم می‌شود
<code>major.minor.build.revision
</code>
می‌توانید شماره نسخه را به صورت زیر مشخص کنید:
<code>[assembly: AssemblyVersion ("2.5.6.7")]</code>
</li></ul></li>

<li><p><strong>فرهنگ (Culture)</strong>:</p><ul><li>فرهنگ از ویژگی AssemblyCulture به دست می‌آید و به اسمبلی‌های ماهواره‌ای اعمال می‌شود. این ویژگی برای اسمبلی‌هایی که برای زبان‌ها و فرهنگ‌های مختلف تولید می‌شوند، مورد استفاده قرار می‌گیرد.</li></ul></li>
<li><p><strong>توکن کلید عمومی (Public Key Token)</strong>:</p><ul><li>توکن کلید عمومی از نام قوی که در زمان کامپایل ارائه می‌شود به دست می‌آید. اگر اسمبلی به صورت Strongly Named (نام قوی) باشد، توکن کلید عمومی آن را شامل می‌شود.</li></ul></li>

</ol>

<h5>مثال توضیحات</h5>
<ul><li><strong>نام ساده</strong>: نام اصلی فایل اسمبلی که به آن کامپایل شده است. برای مثال، اگر فایل اسمبلی <code>MyApp.dll</code> باشد، نام ساده آن "MyApp" خواهد بود.</li><li><strong>نسخه</strong>: نشان‌دهنده نسخه‌ای از اسمبلی است که استفاده می‌شود. به عنوان مثال، <code>[assembly: AssemblyVersion("1.0.0.0")]</code> نشان‌دهنده نسخه 1.0.0.0 از اسمبلی است.</li><li><strong>فرهنگ</strong>: برای اسمبلی‌های محلی‌سازی شده استفاده می‌شود. اگر یک اسمبلی برای فرهنگ خاصی تولید شود، ویژگی AssemblyCulture مقدار مربوط به آن فرهنگ را شامل می‌شود.</li><li><strong>توکن کلید عمومی</strong>: یک هش از کلید عمومی که به اسمبلی Strongly Named اختصاص داده شده است. اگر یک اسمبلی Strongly Named باشد، توکن کلید عمومی بخشی از هویت آن را تشکیل می‌دهد.</li></ul>
</div>

### Fully Qualified Names

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">
<p>یک نام کاملاً واجد شرایط اسمبلی (Fully Qualified Assembly Name) رشته‌ای است که شامل همه چهار جزء هویتی اسمبلی می‌باشد. این چهار جزء عبارتند از: نام ساده (simple name)، نسخه (version)، فرهنگ (culture)، و توکن کلید عمومی (public key token). قالب این نام به صورت زیر است:</p>
</div>

```python
simple-name, Version=version, Culture=culture, PublicKeyToken=public-key
```

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">
<p>به عنوان مثال، نام کاملاً واجد شرایط برای اسمبلی <code>System.Private.CoreLib.dll</code> به صورت زیر است:</p>
</div>

```python
System.Private.CoreLib, Version=4.0.0.0, Culture=neutral, PublicKeyToken=7cec85d7bea7798e
```

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">
<ul><li>اگر اسمبلی ویژگی <code>AssemblyVersion</code> نداشته باشد، نسخه به صورت <code>0.0.0.0</code> نمایش داده می‌شود.</li><li>اگر اسمبلی امضا نشده باشد، توکن کلید عمومی به صورت <code>null</code> نمایش داده می‌شود.</li></ul>

<h5>استفاده از Fully Qualified Names</h5>
<ul><li>خاصیت <code>FullName</code> یک شیء <code>Assembly</code> نام کاملاً واجد شرایط آن اسمبلی را برمی‌گرداند.</li><li>کامپایلر همیشه از نام‌های کاملاً واجد شرایط استفاده می‌کند تا مراجع اسمبلی را در مانیفست (manifest) ضبط کند. این امر تضمین می‌کند که هر اسمبلی به درستی شناسایی و از تداخل با سایر اسمبلی‌ها جلوگیری شود.</li></ul>
</div>

### The AssemblyName Class

`AssemblyName` has two purposes:  
- It `parses` or `builds` a ***fully qualified assembly name***.
  
- It `stores some extra data` to assist in resolving (finding) the assembly.

You can ***obtain an AssemblyName*** object in any of the following ways:
- `Instantiate` an ***AssemblyName***, providing a fully qualified name.
  
- `Call GetName` on an ***existing Assembly***.
- `Call AssemblyName.GetAssemblyName`, providing the `path to an assembly file` on disk.

You can also ***instantiate an AssemblyName*** object `without any arguments`, and then ***set each of its properties***
