This chapter concentrates on the types in the `System.IO` namespace

The .NET ***stream architecture*** centers on **three concepts**: `backing stores`, `decorators`, and `adapters`,

A ***backing store*** is the `endpoint` that makes `input` and `output` useful, such as a **file** or **network connection**. it is either or both of the following  

- A **source** from which `bytes` can be **sequentially** `read`
- A **destination** to which `bytes` can be **sequentially** `written`

A ***Stream*** is a standard set of methods for `reading`, `writing`, and `positioning` for works with **backing stores**.

<p>بر خلاف آرایه‌ها که همه داده‌های پشتیبان آن‌ها به‌صورت هم‌زمان در حافظه وجود دارد، استریم با داده‌ها به‌صورت سریالی (یعنی یک بایت در هر بار یا در بلوک‌هایی با اندازه مدیریت‌پذیر) سر و کار دارد.</p>

***Streams*** fall into two categories:  
- **Backing store streams** : These are `hardwired(مستقیم)` to a particular type of `backing store`, such as `FileStream` or `NetworkStream`.
- **Decorator streams** : <span dir ="rtl">این استریم‌ها از استریم دیگری تغذیه می‌شوند و داده‌ها را به نوعی تغییر می‌دهند، مانند DeflateStream یا CryptoStream.<span>

<div dir="rtl" style="width:90%; margin:auto">
<p>استریم‌های دکوریتور دارای مزایای معماری زیر هستند:</p>
<ul><li>این استریم‌ها Backing store streams را از نیاز به پیاده‌سازی ویژگی‌هایی مانند فشرده‌سازی و رمزنگاری آزاد می‌کنند.</li><li>با دکوریت شدن، رابط (interface) استریم‌ها تغییری نمی‌کند.</li><li>شما می‌توانید دکوریتورها را در زمان اجرا متصل کنید.</li><li>می‌توانید دکوریتورها را به هم زنجیر کنید (مثلاً یک فشرده‌کننده و سپس یک رمزنگار).</li></ul>
</div>

***applications*** often work at **higher levels**, such as `text` or `XML`. Both **backing store** and **decorator** streams deal exclusively in `bytes`.  
***Adapters*** `bridge this gap` by **wrapping** a stream in a class with `specialized methods` typed to a `particular format`.  
  
For example, a ***text reader*** exposes a `ReadLine` method.  
an ***XML writer*** exposes a `WriteAttributes` method.
