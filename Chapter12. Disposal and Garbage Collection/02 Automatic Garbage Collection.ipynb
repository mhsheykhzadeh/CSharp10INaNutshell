```python
public void Test()
{
    byte[] myArray = new byte[1000];

}
```

When ***Test*** `executes`, an array to hold `1,000 bytes` is `allocated` on the `memory heap`.

The array is ***referenced*** by the variable `myArray`, stored on the `local variable stack`.

When the ***method*** `exits`, this local variable `myArray pops out `of scope, `nothing` is left to `reference` the `array` on the `memory heap`.

The ***orphaned array*** then becomes `eligible(واجد شرایط)` to be reclaimed in **garbage collection**

<div dir="rtl" style="width:90%; margin:auto;">
<h3>عملکرد Garbage Collection</h3>
<p>Garbage Collection در CLR به منظور مدیریت حافظه و آزادسازی حافظه اشیاء بلااستفاده (orphaned) طراحی شده است. این فرآیند به صورت خودکار و بهینه‌سازی شده انجام می‌شود، اما به صورت لحظه‌ای و فوری نیست. به جای این، جمع‌آوری زباله به صورت دوره‌ای انجام می‌شود، اما برخلاف جمع‌آوری زباله در خیابان، برنامه زمانی ثابتی ندارد.</p>

<h3>تصمیم‌گیری برای جمع‌آوری زباله</h3>

<p>CLR بر اساس چندین عامل تصمیم می‌گیرد که چه زمانی باید جمع‌آوری زباله را انجام دهد. این عوامل شامل موارد زیر می‌شوند:</p>

<ol><li><strong>حافظه موجود:</strong> اگر حافظه موجود سیستم کم باشد، CLR ممکن است تصمیم بگیرد که جمع‌آوری زباله را زودتر انجام دهد تا فضای حافظه آزاد شود.</li><li><strong>میزان تخصیص حافظه:</strong> اگر برنامه در حال تخصیص مقدار زیادی حافظه باشد، جمع‌آوری زباله ممکن است زودتر اتفاق بیفتد.</li><li><strong>زمان از آخرین جمع‌آوری:</strong> CLR زمان از آخرین جمع‌آوری زباله را نیز در نظر می‌گیرد. اگر زمان زیادی از آخرین جمع‌آوری گذشته باشد، ممکن است زمان مناسبی برای یک جمع‌آوری جدید باشد.</li><li><strong>الگوهای دسترسی به حافظه:</strong> CLR بهینه‌سازی‌هایی انجام می‌دهد تا جمع‌آوری زباله را بر اساس الگوهای دسترسی به حافظه برنامه انجام دهد، به طوری که کارایی برنامه بهبود یابد.</li></ol>

<h3>تأخیر در جمع‌آوری زباله</h3>
<p>به دلیل اینکه تصمیم‌گیری برای جمع‌آوری زباله بر اساس عوامل متعددی انجام می‌شود، بین زمانی که یک شیء بلااستفاده می‌شود و زمانی که حافظه آن واقعاً آزاد می‌شود، یک تأخیر نامشخص وجود دارد. این تأخیر می‌تواند از نانوثانیه‌ها تا چند روز متغیر باشد.</p>

### Roots

A ***root*** is something that `keeps an object alive`.  
  
A **root** is one of the following:
- A `local variable` or `parameter` in an `executing method` (or in any method in its call stack)
- A `static` variable 
- An object on the `queue` that stores objects `ready for finalization` (see the next section)

<div dir="rtl" style="width:90%; margin:auto;">
<p>یک گروه از اشیاء که به طور چرخه‌ای به یکدیگر ارجاع می‌دهند، بدون داشتن یک ریشه مرجع، به عنوان اشیاء مرده در نظر گرفته می‌شوند. به عبارت دیگر، اگر اشیاء نمی‌توانند با دنبال کردن فلش‌ها (مراجع) از یک شیء ریشه‌ای دسترسی پیدا کنند، آن‌ها غیرقابل دسترسی هستند و بنابراین واجد شرایط جمع‌آوری زباله می‌شوند.</p>
