<h3 >1. شروع یک برنامه</h3>
<p><span>وقتی یک برنامه را اجرا می‌کنید، سیستم‌عامل یک process جدید ایجاد می‌کند. این process شامل کد برنامه، داده‌ها، و منابعی مانند فایل‌ها و حافظه است. سیستم‌عامل به این process یک شناسه منحصر به فرد (PID) اختصاص می‌دهد.</span></p>
<h3 >2. ایجاد Thread اصلی</h3>
<p>هنگامی که process ایجاد شد، یک thread اصلی (main thread) نیز ایجاد می‌شود. این thread مسئول اجرای کد اصلی برنامه است. در سی شارپ، این thread معمولاً با متد Main شروع می‌شود.</p>
<h3>3. ایجاد Thread های اضافی</h3>
<p>در طول اجرای برنامه، ممکن است نیاز به ایجاد thread های اضافی داشته باشید تا وظایف مختلف را به صورت همزمان انجام دهید. برای ایجاد یک thread جدید در سی شارپ، می‌توانید از کلاس Thread استفاده کنید. </p>
<h3>4. زمان‌بندی و تخصیص منابع</h3>
<p>سیستم‌عامل وظیفه زمان‌بندی و تخصیص منابع به process ها و thread ها را بر عهده دارد. سیستم‌عامل تصمیم می‌گیرد که کدام thread روی کدام هسته اجرا شود و چگونه منابع بین thread ها و process ها توزیع شود. این فرآیند به نام زمان‌بندی (scheduling) شناخته می‌شود.</p>

<h3>CPU تک هسته‌ای</h3>
<p>در یک <code>CPU</code> تک هسته‌ای، هر هسته در هر لحظه تنها می‌تواند یک <code>thread</code> را اجرا کند. اما سیستم‌عامل به سرعت بین <code>thread</code> های مختلف جابجا می‌شود (زمان‌بندی چندوظیفه‌ای یا <code>time-slicing</code>)، به طوری که به نظر می‌رسد همه thread ها به صورت همزمان در حال اجرا هستند. این جابجایی به قدری سریع است که کاربران متوجه نمی‌شوند که در هر لحظه تنها یک thread در حال اجرا است.</p>
<h3>CPU چند هسته‌ای</h3>
<p>در یک CPU چند هسته‌ای، هر هسته می‌تواند به طور مستقل یک thread را اجرا کند. بنابراین، اگر یک CPU دارای چندین هسته باشد، می‌تواند به طور واقعی چندین thread را به صورت همزمان اجرا کند. این به معنای افزایش کارایی و کاهش زمان اجرای برنامه‌ها است، زیرا هر هسته می‌تواند به طور مستقل وظایف خود را اجرا کند.</p>

<p>این تفاوت‌ها باعث می‌شود که در یک CPU چند هسته‌ای، کارها به صورت واقعی همزمان انجام شوند، در حالی که در یک CPU تک هسته‌ای، این همزمانی به صورت مجازی و با جابجایی سریع بین thread ها اتفاق می‌افتد.</p>
