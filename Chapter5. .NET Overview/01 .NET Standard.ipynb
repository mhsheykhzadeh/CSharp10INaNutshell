***.NET Standard*** is not a `runtime`;

### .NET Standard 2.0

The most useful version is` .NET Standard 2.0`. It will run without modification on most `modern` and `legacy` ***runtimes*** that are still in use today, including:  
- `.NET Core 2.0+` (including .NET 5 and .NET 6)
- `UWP 10.0.16299+`
- `Mono 5.4+` (the CLR/BCL used by older versions of Xamarin)
- `.NET Framework 4.6.1+`

```python
<!-- target .NET Standard 2.0, add the following to your .csproj file -->
<PropertyGroup>
    <TargetFramework>netstandard2.0</TargetFramework>
<PropertyGroup>
```

### .NET Standard 2.1

NET Standard 2.1 is a `superset` of ***.NET Standard 2.0*** that supports (only) the
following platforms:  
- .NET Core 3+
- Mono 6.4+

The following APIs, in particular, are available in .NET Standard 2.1 (but not .NET
Standard 2.0):
- Span<T> (Chapter 23)
- Reflection.Emit (Chapter 18)
- ValueTask<T> (Chapter 14)

<h3>مشکل وجود کتابخانه‌های قدیمی</h3>
<ul><li><strong>کتابخانه‌های قدیمی:</strong> از آنجا که .NET Framework برای مدت طولانی وجود داشته است، بسیاری از کتابخانه‌های موجود فقط برای .NET Framework توسعه یافته‌اند و نسخه‌ای برای .NET Standard، .NET Core، یا .NET 6 ندارند.</li><li><strong>مشکل عدم سازگاری:</strong> این مسئله باعث می‌شود که استفاده از این کتابخانه‌ها در پروژه‌های جدیدتر مشکل باشد.</li></ul>
<h3>راه‌حل در پروژه‌های جدیدتر</h3>
<ul><li><strong>مجاز بودن استفاده از کتابخانه‌های .NET Framework:</strong> در پروژه‌های .NET 5+ و .NET Core، امکان استفاده از کتابخانه‌های .NET Framework وجود دارد. این به توسعه‌دهندگان اجازه می‌دهد تا از کتابخانه‌های قدیمی در پروژه‌های جدید استفاده کنند.</li><li><strong>شرایط استفاده:</strong> با این حال، چند شرط وجود دارد:<ul><li><strong>استثنائات API:</strong> اگر کتابخانه .NET Framework یک API را فراخوانی کند که در .NET Core یا .NET 5+ پشتیبانی نمی‌شود، یک استثنا (exception) ایجاد می‌شود.</li><li><strong>وابستگی‌های پیچیده:</strong> وابستگی‌های غیرساده (nontrivial) ممکن است (و اغلب هم می‌شود) نتوانند به درستی حل شوند. به عبارت دیگر، اگر کتابخانه‌ای وابستگی‌های پیچیده‌ای داشته باشد، ممکن است نتواند به درستی در پروژه‌های جدیدتر کار کند.</li></ul></li></ul>
<h3>کاربرد عملی</h3>
<ul><li><strong>موارد ساده:</strong> در عمل، این روش بیشتر در موارد ساده کار می‌کند. به عنوان مثال، اگر یک کتابخانه فقط یک DLL غیرمدیریتی (unmanaged DLL) را بسته‌بندی کرده باشد، احتمال بیشتری دارد که به درستی کار کند.</li></ul>
